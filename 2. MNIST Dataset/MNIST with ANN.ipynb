# MNIST DataSet Project

### Importing Libraries

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

### Loading Data

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

### Exploring Data Shapes

```python
X_train.shape
```

### Exploring Test Data

```python
X_test[0]
```

```python
y_test[0]
```

### Visualizing a Sample Image

```python
# Use of imshow()

import matplotlib.pyplot as plt
plt.imshow(X_test[0])
```

### Scaling Values

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

### Building the Model

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
# model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

### Compiling the Model

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

### Training the Model

```python
history = model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

### Checking Accuracy_Score

```python
# Because of Softmax -> we'll Get o/p in prob.

y_prob = model.predict(X_test)
```

```python
# argmax tells about the max probabilty

y_pred = y_prob.argmax(axis=1)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### Drawing Graph

```python
plt.plot(history.history['loss'],label='Loss_line')
plt.plot(history.history['val_loss'],label='Val_Loss_line')
plt.legend()
```

```python
plt.plot(history.history['accuracy'],label='Accuracy_line')
plt.plot(history.history['val_accuracy'],label='Val_Accuracy_line')
plt.legend()
```
